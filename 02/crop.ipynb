```python
import os
from PIL import Image

def splitimage(src,rownum, colnum, dspath):
    img = Image.open(src)
    w, h = img.size
    if rownum <= h and colnum <= w:
        print('Original image info: %sx%s, %s, %s' % (w, h, img.format, img.mode))
        print('图片切割')

        s = os.path.split(src)
        if dspath == '':
            dspath = s[0]
        fn = s[1].split('.')
        basename = fn[0]
        ext = fn[-1]

        num = 0
        rowheight = h // rownum
        colwidth = w // colnum
        for r in range(rownum):
            for c in range(colnum):
                box = (c*colwidth, r*rowheight, (c+1)*colwidth, (r+1)*rowheight)
                img.crop(box).save(os.path.join(dspath, basename + str(num)+'.'+ 'jpeg'), 'jpeg')
                num = num + 1
        print('共生成%s图' % num)
    else:
        print('error')

def mkdir(path):
    path = path.strip()
    path = path.rstrip('\\')
    isExists = os.path.exists(path)

    if not isExists:
        os.makedirs(path)
        print(path + 'success')
        return True

    else:
        print(path + 'error')
        return False

folder = r'D:/code/cave/watercolors'
path = os.listdir(folder)

for each_jpg in path:
    first_name, second_name = os.path.splitext(each_jpg)
    each_jpg = os.path.join(folder, each_jpg)
    src = each_jpg
    print(src)
    print(first_name)
    mkpath = "D:/code/results/watercolors/"+ first_name
    mkdir(mkpath)
    if os.path.isfile(src):
        dspath = mkpath
        if(dspath == '') or os.path.exists(dspath):
            row = int(2) # 切割行数
            col = int(2) # 切割列数
            if row > 0 and col > 0:
                splitimage(src, row, col, dspath)
            else:
                print('无效的')
        else:
            print('图片保存目录 %s 不存在！' % dspath)
    else:
        print('图片文件 %s 不存在！' % src)
```
